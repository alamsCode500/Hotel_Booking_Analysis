## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Fill null values of missing values columns
try:
    # Code that may raise an exception

    # Make copy of dataframe
    df_hotel = df_hotel_booking.copy()

    # Fill missing values in 'company', 'agent' and in 'country' with 'Unknown'
    df_hotel['company'].fillna('Unknown', inplace=True)
    df_hotel['agent'].fillna('Unknown', inplace=True)
    df_hotel['country'].fillna('Unknown', inplace=True)

    # Filling missing values in the 'children' column with the mean value of the column
    mean_children = df_hotel['children'].mean()
    #print(mean_children)
    df_hotel['children'].fillna(mean_children, inplace=True)

    # Check null values are filled
    print(df_hotel.isnull().sum())

except Exception as e:
    # Handle any type of exception
    print("An error occurred:", e)
else:
    # Run if no exception occurred
    print("No exception occurred")
finally:
    # Always executed, whether an exception occurred or not
    print("Execution of the try-except block is complete")
````

````python
# Convert Data Types of the columns
try:
    # Code that may raise an exception

    # Check the data type of reservation_status_date column
    print("\nBefore: reservation_status_date column datatype \n", df_hotel['reservation_status_date'].dtype)

    # Convert 'reservation_status_date' column from object to datetime
    df_hotel['reservation_status_date'] = pd.to_datetime(df_hotel['reservation_status_date'])

   # Check the data type of reservation_status_date column
    print("\nAfter: reservation_status_date column datatype \n", df_hotel['reservation_status_date'].dtype)

    # Convert relevant categorical columns to 'category' data type for efficiency
    categorical_cols = [
        'hotel', 'meal', 'country', 'market_segment',
        'distribution_channel', 'customer_type',
        'reservation_status', 'reserved_room_type',
        'assigned_room_type', 'deposit_type'
    ]

    # Check the data type of categorical_cols columns list
    print("\nBefore: categorical_cols columns datatypes \n", df_hotel[categorical_cols].dtypes)

    # convert object datatype columns to category datatype
    df_hotel[categorical_cols] = df_hotel[categorical_cols].astype('category')

    # Check the data type of categorical_cols columns list
    print("\nAfter: categorical_cols columns datatypes \n", df_hotel[categorical_cols].dtypes)

except Exception as e:
    # Handle any type of exception
    print("An error occurred:", e)
else:
    # Run if no exception occurred
    print("No exception occurred")
finally:
    # Always executed, whether an exception occurred or not
    print("Execution of the try-except block is complete")
````

````python
# Create New columns from the given columns of the dataset
try:
    # Code that may raise an exception

    # Calculate total guests in each record
    df_hotel['total_guests'] = df_hotel['adults'] + df_hotel['children'] + df_hotel['babies']

    # Calculate total duration of stay in hotel
    df_hotel['total_stay'] = df_hotel['stays_in_weekend_nights'] + df_hotel['stays_in_week_nights']

    # Create a column for whether the booking was made by a repeated guest
    print("Before: is_repeated_guest datatype :: ", df_hotel['is_repeated_guest'].dtype)
    df_hotel['is_repeated_guest'] = df_hotel['is_repeated_guest'].astype(bool)
    print("After: is_repeated_guest datatype :: ", df_hotel['is_repeated_guest'].dtype)

    # Create a new column for the total number of bookings made by the guest
    df_hotel['total_bookings'] = df_hotel['previous_bookings_not_canceled'] + df_hotel['previous_cancellations']

    # print first five rows of new columns
    print(df_hotel[['total_guests', 'total_stay', 'is_repeated_guest', 'total_bookings']].head())

except Exception as e:
    # Handle any type of exception
    print("An error occurred:", e)
else:
    # Run if no exception occurred
    print("No exception occurred")
finally:
    # Always executed, whether an exception occurred or not
    print("Execution of the try-except block is complete")
````

````python
# Chart - 18 visualization code

# :: Multivariate Analysis (More than two variables) ::

try:
    # Code that may raise an exception

    # Heatmap of Average Daily Rate and Cancellation Rate by Month

    # Get month name in month column
    df_hotel['month'] = df_hotel['reservation_status_date'].dt.month_name()

    # Group by month then calculate mean of adr and is_canceled by using aggregate function
    monthly_stats = df_hotel.groupby('month').agg({'adr': 'mean', 'is_canceled': 'mean'}).reset_index()

    # Figure size 12 on x-axis and 6 on y-axis
    plt.figure(figsize=(12, 6))

    # Heat Map on pivot table
    sns.heatmap(monthly_stats.pivot(index = 'month', columns = 'adr', values = 'is_canceled'), cmap='YlGnBu', annot=True)

    # Print the title
    plt.title('Average Daily Rate and Cancellation Rate by Month')

    # show the plot
    plt.show()

except Exception as e:
    # Handle any type of exception
    print("An error occurred:", e)
else:
    # Run if no exception occurred
    print("No exception occurred")
finally:
    # Always executed, whether an exception occurred or not
    print("Execution of the try-except block is complete")
````
